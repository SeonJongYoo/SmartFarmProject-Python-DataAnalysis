```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import bs4
import pandas as pd
```

```python
keyword = '셀러리'
driver = webdriver.Chrome(r'C:\Users\jobgs\Desktop\CONCAT_SMARTFARM_Project\DataScience\SMARTFARM_Project-DataAnalysis_Plant\chromedriver.exe')
driver.get('https://search.shopping.naver.com/search/all?query={}&cat_id=&frm=NVSHATC'.format(keyword))
```

```python
time.sleep(5)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
time.sleep(1)


item_source = bs4.BeautifulSoup(driver.page_source, 'lxml')
item_list = item_source.select('div.basicList_info_area__17Xyo')
item_list
```

```python
title_list = []
price_list = []
review_list = []
buy_list = []

for item in item_list:
    title = item.select_one('a').text.replace('\n', '')
    price = item.select_one('span.price_num__2WUXn').text
    ReviewAndBuy = item.select('a.basicList_etc__2uAYO > em')
    try:
        review = ReviewAndBuy[0].text + '개'
    except IndexError:
        review = '없음'
    
    try:
        buy = ReviewAndBuy[1].text + '개'
    except IndexError:
        buy = '없음'
        
    title_list.append(title)
    price_list.append(price)
    review_list.append(review)
    buy_list.append(buy)
#print(review_list)
#print(title_list, price_list, review_list, buy_list)
df = pd.DataFrame({'상품명':title_list, '가격':price_list, '리뷰수':review_list, '구매건수':buy_list})
```

```python
df
```

```python
gram_tmp = df.상품명.str.extract('(\d+g|\d+kg|\d+Kg|\d+KG)')
df['무게당 가격'] = gram_tmp.dropna()
df1 = df.dropna()
df1.reset_index(inplace=True)
del df1['index']
df1
```

```python
for i in range(len(df1)):
    tmp = df1.loc[i][0]
    if '즙' in tmp or '주스' in tmp or '쥬스' in tmp or '팩' in tmp:
        print(df1.index[i])
        df1.drop(df1.index[i], inplace=True)
```

```python
# 무게만 추출
gram_list = []
for i in range(len(df1)):
    tmp = df1.loc[i][4]
    if 'kg' in tmp:
        wei = tmp.split('kg')
        res = int(wei[0])*1000
    elif 'Kg' in tmp:
        wei = tmp.split('Kg')
        res = int(wei[0])*1000
    elif 'KG' in tmp:
        wei = tmp.split('KG')
        res = int(wei[0])*1000
    elif 'g' in tmp:
        wei = tmp.split('g')
        res = int(wei[0])
    gram_list.append(res)
print(gram_list)
```

```python
# 가격 추출
plist = []
for i in range(len(df1)):
    tmp = df1.loc[i][1]
    tmp1 = tmp.split('원')
    tmp2 = tmp1[0].split(',')
    tmp3 = ''.join(tmp2)
    plist.append(int(tmp3))
```

```python
# 그램당 가격 추출하기
GramPerPrice = []
for i in range(len(df1)):
    res = plist[i] // gram_list[i]
    GramPerPrice.append(str(res)+'원/g')
df1['가격/g'] = GramPerPrice
del df1['무게당 가격']
df1
```
